```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train_data = matches_info[matches_info["date"] < '2022-01-01']
```

```python
test_data = matches_info[matches_info["date"] > '2022-01-01']
```

```python
predictors = ["venue_int", "opponent_int", "hour", "day_int"]
```

```python
rf.fit(train_data[predictors], train_data["target"])
```

```python
preds = rf.predict(test_data[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(test_data["target"], preds))
```

```python
combined = pd.DataFrame(dict(actual=test_data["target"], predicted=preds))
```

```python
from sklearn.metrics import precision_score

precision_score(test_data["target"], preds)
```

```python
grouped_matches = matches_info.groupby("team")
```

```python
group = grouped_matches.get_group("Arsenal").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk","pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches_info.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```
